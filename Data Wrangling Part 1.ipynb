Merging these datasets will be based on stop_id because this uniquely identifies each stop, whereas pid uniquely identifies individuals in each stop. 

Since there may be multiple individuals per stop_id, we will have multiple observations per stop after the merge 

```python
#conduct a inner merge because we want to keep all stop details located in ripa_basic &results 
    #merge is to add additional information not in ripa_basic to those stops 

#keys for merge are stop_id, and pid 

ripa_basic_result = pd.merge(ripa_basic, result, on=['stop_id','pid'])
ripa_basic_result.head()

print(ripa_basic_result.shape) #duplicated
```

```python
ripa_basic_result['stop_id'].nunique()
```

```python
ripa_basic['stop_id'].nunique()
```

```python
refined_basic = ripa_basic.drop_duplicates(subset=['stop_id','pid'])
refined_results = result.drop_duplicates(subset = ['stop_id','pid'])

basic_result_combo = pd.merge(refined_basic, refined_results, how='inner', left_on=['stop_id','pid'],right_on=['stop_id','pid'])
print(refined_basic.shape)
print(refined_results.shape)
print(basic_result_combo.shape)
```

Now to merge on other dataframes: race, reason, and actions taken 
First remove any duplicates based on stop_id and pid, then conduct merge

```python
refined_race = race.drop_duplicates(subset=['stop_id','pid'])

refined_reason = reason.drop_duplicates(subset=['stop_id','pid'])

refined_actions = actions_taken.drop_duplicates(subset=['stop_id','pid'])

refined_contraband = contraband.drop_duplicates(subset = ['stop_id','pid'])
```

```python
df_merge = [basic_result_combo,refined_race, refined_reason, refined_actions, refined_contraband]

from functools import reduce #required for easier merging all the dataframes at once

ripa_all_combo = reduce(lambda left, right: pd.merge(left, right, on =['stop_id','pid']),df_merge)

print(ripa_all_combo.shape)

#ripa_all_combo = pd.merge(basic_result_combo, refined_results, how='inner', on=['stop_id','pid'])
```
